# Importing Files - **SINDH**

```python
try:
    df1, meta1 = pyreadstat.read_sav('/content/fs-sindh.sav')
    print("df loaded successfully using pyreadstat.")
    for col in df1.columns:
        print(col)

    df2, meta2 = pyreadstat.read_sav('/content/hh-sindh.sav')
    print("df loaded successfully using pyreadstat.")
    for col in df2.columns:
        print(col)

except Exception as e:
    print(f"Error loading fs-sindh.sav using pyreadstat: {e}")
```

```python
merged_df1 = pd.merge(df1, df2, on=['HH1', 'HH2'], how='inner')
print("Merged DataFrame info:")
merged_df1.info()
print("\nMerged DataFrame head:")
display(merged_df1.head())

merged_df1.to_csv("merged_df1.csv", index=False)
```

```python
# Make a copy of the merged DataFrame
copied_merged_df1 = merged_df1.copy()

print("Copied DataFrame info:")
copied_merged_df1.info()
print("\nCopied DataFrame head:")
display(copied_merged_df1.head())
```

# ***cleaning***

```python
# Define the list of columns to keep
columns_to_keep = [
    'HH1', 'HH2',  # Merge keys
    'FSAGE_x',  # Age of child (continuous)
    'FS2_x',    # Gender of child (male/female dummy, assuming 1=male, 2=female) - need to confirm mapping
    'HH52_x',# Birth order / number of siblings (optional) - Need to identify this variable
    'melevel_x', # Mother's education (categorical)
    'hh7r_y',   # Household Head's education (categorical)
    'windex5_y', # Household wealth index quintile
    'HH6_y',    # Household size (total members)
    'HH4_y',    # Head of household's sex (female-headed vs male-headed)
    'CB7_x',  #(Attended school or early childhood programme during current school year)
    'CB4_x',  #(Ever attended school): Broader measure; can be used for robustness check.
    'CB9_x',  # (Attended during previous year): Can show persistence
]

# Filter the DataFrame to keep only the specified columns
cleaned_df1 = copied_merged_df1[columns_to_keep].copy()

print("Cleaned DataFrame info:")
cleaned_df1.info()
print("\nCleaned DataFrame head:")
display(cleaned_df1.head())
```
